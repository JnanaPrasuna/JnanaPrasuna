**EDA Top 100 YouTube Channels in 2024**

This dataset showcases the top 100 YouTube channels in 2024, including key metrics such as subscribers, total views, and video count. It has been manually curated to provide a snapshot of the leading channels as of December 2024. The data reflects the performance of channels across various genres, from entertainment to educational content, offering insights into the most influential creators of the year. This data is useful for marketers, content creators, and data scientists who want to analyze trends, audience engagement, and growth strategies of top YouTube influencers. By examining these metrics, users can gain valuable insights into content creation, viewer preferences, and platform dynamics. The dataset is based on publicly available YouTube statistics and third-party platforms, ensuring transparency and accuracy.

```python
# Essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/YOUTUBE CHANNELS DATASET.csv')
```

```python
#Dropping unneccesary columns
df = df.drop(columns=[col for col in df.columns if "Negative" in col],axis=1)
```

```python


#check first five rows
df.head()

#check missing values
df.isnull().sum()

#Statistics of numeric columns
df.describe()

#check data type of columns
df.dtypes
```

```python
#plot country-wise frequency from existing dataframe
df['Country'].value_counts().plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title('Country-wise Frequency of YouTube Channels')
plt.xlabel('Country')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45)
plt.tight_layout()
```

The code counts the frequency of each unique value in the "Country" column of the DataFrame df using value_counts(). It then plots the frequencies as a bar chart, customizing the appearance with a sky-blue color and rotating the x-axis labels for better readability. Finally, it adjusts the layout and displays the plot using plt.show().

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', xticklabels=df.columns, yticklabels=False)
plt.title('Heatmap of Missing Values')
plt.tight_layout()
plt.show()
```

The heatmap visually represents missing values in the dataset, where True (yellow) indicates missing data and False (purple) shows non-missing data. It helps to easily identify columns and rows with missing values for further analysis.

```python
# Plot boxplot for 'Subscribers' column
sns.boxplot(x=df['Subscribers'], color='lightblue')

# Add title and display the plot
plt.title('Boxplot of Subscribers')
plt.tight_layout()
plt.show()
```


The boxplot for the Subscribers column displays the distribution of subscriber counts for the YouTube channels. It shows the median (the line inside the box), the interquartile range (the box itself), and potential outliers (points outside the whiskers). The boxplot helps identify the central tendency, variability, and any extreme values in the Subscribers data, providing insights into how subscribers are distributed across the channels.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create scatter plot for 'Country' vs 'Subscribers'
sns.scatterplot(data=df, x='Country', y='Subscribers', hue='Country', palette='Set2')

# Add title and labels
plt.title('Scatter Plot of Country vs Subscribers')
plt.xlabel('Country')
plt.ylabel('Subscribers')
plt.xticks(rotation=90)  # Rotate country labels if they overlap
plt.tight_layout()
plt.show()


```

The scatter plot shows the relationship between the Country and the number of Subscribers for each YouTube channel. Each point represents a channel, and the points are colored by country, using a distinct palette. The plot helps visualize how subscriber counts are distributed across different countries, with the x-axis representing countries and the y-axis representing subscriber counts.

```python
# Create a line plot for 'Country' vs 'Subscribers'
sns.lineplot(data=df, x='Country', y='Subscribers', marker='o', color='blue')

# Add title and labels
plt.title('Line Plot of Country vs Subscribers')
plt.xlabel('Country')
plt.ylabel('Subscribers')
plt.xticks(rotation=90)  # Rotate country labels if they overlap
plt.tight_layout()
plt.show()
```


The line plot shows the relationship between Country and Subscribers, with each point representing the number of subscribers for a particular country. The x-axis displays the countries, while the y-axis shows the subscriber count. The plot uses lines to connect the points, with markers highlighting individual data points. This visualization helps in observing trends in subscribers across different countries, although line plots are typically more effective for continuous data rather than categorical variables like Country.
