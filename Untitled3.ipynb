Basic Line Plot

Plot a simple line graph of ùë¶ = ùë• 2 for ùë• values ranging from -10 to 10. Label the x-axis as "X Values" and the y-axis as "Squared Values." Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-10,11)
y=x**2
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Simple Line Plot")
plt.show()
```


Bar Plot


Scatter plot

Generate two lists of 20 random integers each (for x and y coordinates). Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
x=[np.random.randint(1,100) for i in range(20)]
y=[np.random.randint(1,100) for i in range(20)]
plt.scatter(x,y,c='purple')
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.title("Scatter Plot of Random Points")
plt.show()
```


Histogram


Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram. Customize the number of bins and add a grid to the plot.

```python


data=np.random.randn(100)
plt.hist(data,bins=10)
plt.grid(True)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Random Values")
plt.show()

```


Pie Chart

Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10]. Display the percentage values on the chart and use a color palette for differentiation.


```python


categories=['A','B','C','D','E']
values=[20,30,25,15,10]
colors=['red','green','blue','yellow','orange']
plt.pie(values,labels=categories,colors=colors,autopct='%1.1f%%')
plt.title("Percentage Distribution of Categories")
plt.show()
```


Heatmap (Seaborn)


Create a 5x5 matrix with random integers between 1 and 10 using np.random.randint() and visualize it as a heatmap using Seaborn. Add a color bar and annotate the heatmap with the values.


```python


matrix=np.random.randint(1,11,size=(5,5))
sns.heatmap(matrix,annot=True,cmap='coolwarm',cbar=True)
plt.title("Heatmap of Random Integers")
plt.show()
```


Comparison of Matplotlib and Seaborn


Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn. Highlight the differences in the code, syntax, and aesthetics of the two libraries.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns # Import seaborn before using it


students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks scored by 5 Students using matplotlib")
plt.show()

data = pd.DataFrame({'Student': students, 'Marks': marks})
sns.barplot(x='Student', y='Marks', data=data) # Now sns is defined
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks scored by 5 Students using seaborn")
plt.show()

```
